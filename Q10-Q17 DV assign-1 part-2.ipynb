```python
df['Reviews']=df['Reviews'].str.replace('3.0M','3000000')
```

```python
df['Reviews']=df['Reviews'].astype('int')
```

```python
df.info()
```

```python
df.drop(index=10472,inplace=True)
```

```python
df['Size'].unique()
```

```python
def convert_size(size):
  if 'M' in size:
    return int(float(size.replace('M',''))*1000)
  elif 'K' in size:
    return float(size.replace('k',''))
  elif 'Varies with device' in size:
    return np.nan
  else:
    return np.nan
```

```python
df['Size']=df['Size'].apply(convert_size)
```

```python
df['Size'].unique()
```

```python
df.info()
```

```python
df['Installs'].unique()
```

```python
cols=['Installs','Price']
chars_to_remove=['+',',','$']
for i in cols:
  for j in chars_to_remove:
    df[i]=df[i].str.replace(j,'')
```

```python
rows, col = df.shape
```

```python
# question10 :app ratings by category

plt.figure(figsize=(20,10))
sns.boxplot(x=df['Category'],y=df['Rating'])
plt.title('Box Plot: Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Ratings')
plt.show()
```

```python
# Q12 :CREATE A SCATTER PLOT
plt.figure(figsize=(12, 8))
plt.scatter(df['Size'], df['Rating'], alpha=0.5)
plt.title('Scatter Plot of App Size vs Rating')
plt.xlabel('Size (M)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
#Q13 :BAR CHART FOR AVG PRICE vs APP TYPE
# Convert 'Price' column to numeric type
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Calculate average price grouping it up by apps
avg_price = df.groupby('App')['Price'].mean().head(30)

plt.bar(avg_price.index, avg_price.values)
plt.xlabel('Apps')
plt.ylabel('Average price in dollars')
plt.xticks(rotation=90)

plt.show()
```

```python
# Q14:Create a frequency table to identify the apps with the highest number of installs.
freq_table = df.groupby(['App', 'Installs']).size().reset_index(name='Frequency').head(10)
print(freq_table)
```
